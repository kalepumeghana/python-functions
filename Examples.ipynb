```python
def fun():
    name = input("enter your name")
    age = int(input("enter your age"))
    city = input("enter your city")
    course = input("enter your course")
    print("Name:", name)
    print("Age:", age)
    print("city:", city)
    print("course:", course)


while True:
    fun()
    ch = input("do you want to continue?")
    if ch == "no":
        break
```

create a function called tree with parameters monkey_count,fruit_type,fruit count. this function should calculate the time taken by the total monkeys to completely eat the fruits an prints the details.

```python
def tree(monkey_count,fruit_type,fruit_count):
     monkey_count=int(input())
     fruit_type=input()
     fruit_count=int(input())
     d={"mango":2,"apple":3,"orange":1.5}
     t=d[fruit_type.lower()]
     time_taken=int(fruit_count/monkey_count)*t
     print(time_taken)
tree(5,"mango",10)
```
