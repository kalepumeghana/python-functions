```
def multiply(a,b,c):
    print(a*b*c)
multiply(2,3,4)
```

```
def describe_pet(animal,name):
    print(f"My {animal} is named {name}")
describe_pet("dog","cutie")
```

```
def power(base,exponent):
    print(base**exponent)
power(2,5)
```

```
def full_name(first,middle,last):
    print(f"My name is {first}{middle}{last}")
full_name("kalepu",""," Meghana")
```
